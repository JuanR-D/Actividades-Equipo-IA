```python
#Juan Carlos Reyes Delgadillo 2011149
#Walter Aldair Torres Avila 1999749
#Rogelio Jesus Tovar Zepeda 2011149
#Edson de Jesus Olivares Vazquez 2002238
# Cargar librerías
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression

# Cargar datos desde un archivo CSV
df = pd.read_csv('/content/robot_sensors.csv')

# Muestra las primeras líneas de datos.
print(df.head(10))
# Obtén información sobre las columnas y el tipo de datos de cada una.
print(df.info(10))
print(df.shape)

# Verifica si hay valores nulos (missing values) en alguna columna y cómo manejarlos.
print(df.isnull().sum())
# Si encuentras valores nulos, puedes optar por rellenarlos con la media o eliminar las filas que contienen datos faltantes.
df['Distancia (cm)'] = df['Distancia (cm)'].fillna(df['Distancia (cm)'].mean())
df['Velocidad (cm/s)'] = df['Velocidad (cm/s)'].fillna(df['Velocidad (cm/s)'].mean())
df['Temperatura (°C)'] = df['Temperatura (°C)'].fillna(df['Temperatura (°C)'].mean())
#Filtrar valores negativos de la velocidad
df_filtered = df[df['Velocidad (cm/s)'] >= 0]

#Normalizar Datos
# Seleccionar las columnas numéricas a normalizar
numerical_features = ['Distancia (cm)', 'Velocidad (cm/s)', 'Temperatura (°C)']
# Crear un objeto MinMaxScaler
scaler = MinMaxScaler()
# Ajustar el scaler a los datos y transformarlos
df[numerical_features] = scaler.fit_transform(df[numerical_features])
#visualizar los datos obtenidos con matplot lib, ahora plt

#primero distancia
x=('Distancia (cm)')
y=('Tiempo (s)')
plt.scatter(df[x], df[y], color='blue', label="tiempo-distancia")
plt.legend()  # Muestra la leyenda
plt.grid(True)  # Muestra la cuadrícula
plt.title("DISPERSION, distancia-tiempo")  # Título
plt.xlabel("distancia (cm)")  # Etiqueta del eje x
plt.ylabel("tiempo (s)")  # Etiqueta del eje y
plt.show()  # Muestra el gráfico


#luego velocidad
h=('Velocidad (cm/s)')
plt.scatter(df[h], df[y], color='green', label="tiempo-velocidad")
plt.legend()  # Muestra la leyenda
plt.title("Dispersión, velocidad-tiempo")  # Título
plt.xlabel("velovidad(m/s)")  # Etiqueta del eje x
plt.ylabel("tiempo(s)")  # Etiqueta del eje y
plt.grid(True)  # Muestra la cuadrícula
plt.show()


#luego temperatura
i=('Temperatura (°C)')
plt.scatter(df[i], df[y], color='red', label="tiempo-temperatura")
plt.legend()  # Muestra la leyenda
plt.title("Dispersión, temperatura-tiempo")  # Título
plt.xlabel("temperatura(°C)")  # Etiqueta del eje x
plt.ylabel("tiempo(s)")  # Etiqueta del eje y
plt.grid(True)  # Muestra la cuadrícula
plt.show()

```
