```python
#-Walter Aldair Torres Avila 1999749 IMTC
#-Edson de jesus olivares vazquez 2002238
#-Tovar Zepeda Rogelio Jesús 2011149
#-Juan Carlos Reyes Delgadillo 2011149
#GRUPO:007

# Importar las librerías necesarias
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

# --- Carga y Preprocesamiento de Datos ---
print("Cargando y preprocesando datos...")
# Cargar el conjunto de datos MNIST usando la función incorporada de Keras
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Guardar una copia de los datos de prueba originales para visualización posterior
# Esta copia no se aplanará ni normalizará para poder mostrar la imagen visual.
(x_train_original, y_train_original), (x_test_original, y_test_original) = keras.datasets.mnist.load_data()

# Preprocesar los datos para el entrenamiento (normalizar y aplanar)
# Estos datos son los que se usarán para entrenar y evaluar el modelo.
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

x_train = x_train.reshape(x_train.shape[0], 28 * 28)
x_test = x_test.reshape(x_test.shape[0], 28 * 28)

print("Datos cargados y preprocesados.")

# --- Construcción y Entrenamiento del Modelo ---
print("\nConstruyendo y compilando el modelo...")
# Definir la estructura de la red neuronal
model = keras.models.Sequential([
    # Capa de entrada y primera capa densa con activación ReLU
    keras.layers.Dense(128, activation='relu', input_shape=(784,)),
    # Capa de Dropout para regularización
    keras.layers.Dropout(0.2),
    # Capa de salida con 10 unidades (una por cada dígito) y activación Softmax
    keras.layers.Dense(10, activation='softmax')
])

# Compilar el modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("Modelo construido y compilado.")

# Entrenar el modelo
print("\nEntrenando el modelo...")
# Monitoreamos la exactitud y pérdida en un conjunto de validación (10% del entrenamiento)
history = model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1)

print("Entrenamiento completado.")

# --- Evaluación del Modelo ---
print("\nEvaluando el modelo en los datos de prueba...")
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)

# Mostrar la exactitud en los datos de prueba
print('\nExactitud en los datos de prueba:', test_acc)

# --- Predicciones y Visualización ---
print("\nRealizando predicciones y visualizando resultados para las primeras 5 imágenes de prueba...")

# Realizar 5 predicciones con el conjunto de datos de prueba (aplanado y normalizado)
# Usamos x_test que es la data que el modelo espera como entrada.
predictions = model.predict(x_test[:5])

# Mostrar las predicciones y los valores reales
print("\nResultados de las primeras 5 predicciones:")

# Iterar sobre las primeras 5 imágenes
for i in range(5):
    # Obtener la predicción (el índice de la clase con la mayor probabilidad)
    predicted_label = np.argmax(predictions[i])
    # Obtener el valor real (la etiqueta correcta de la imagen)
    # Usamos y_test_original que contiene las etiquetas sin modificar.
    real_label = y_test_original[i]

    # --- Mostrar la imagen original (sin aplanar, remodelada a 28x28 para visualización) ---
    # Usamos x_test_original que no fue aplanado.
    plt.imshow(x_test_original[i], cmap='gray')
    plt.title(f"Imagen Original\nPredicción: {predicted_label}, Real: {real_label}")
    plt.axis('off') # Ocultar los ejes para una mejor visualización
    plt.show()

    # --- Mostrar la representación aplanada (los datos que entraron al modelo) ---
    print(f"Representación Aplanada de la Imagen {i+1} (primeros 50 valores del vector de entrada):")
    # Mostramos los primeros 50 valores del arreglo aplanado y normalizado (x_test).
    print(x_test[i][:50])
    print("...") # Indicamos que hay más valores en el arreglo (total 784)

    # Imprimir el resumen de la predicción y el valor real
    print(f"Imagen {i+1}: Predicción = {predicted_label}, Valor Real = {real_label}")
    print("-" * 30) # Separador para las diferentes imágenes
```
